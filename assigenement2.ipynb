$$

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


## Answer



---

## 1. Probability Space

Consider the probability space

$$
\Omega=[0,1],
\qquad
\mathfrak F=\mathfrak B([0,1]),
$$

where $\mathfrak B([0,1])$ denotes the Borel $\sigma$-algebra on the interval $[0,1]$.

Define the probability measure $P$ by

$$
P(A)=\int_A f(x)\,dx,
\qquad
A\in\mathfrak F,
$$

where the density function is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x(1-x)^{17},
& 0<x<1,\\
0,
& \text{otherwise}.
\end{cases}
$$

---

## 2. Definition of the Random Variable

Define the random variable

$$
X:\Omega\to\mathbb R
$$

by

$$
X(\omega)=\omega,
\qquad
\omega\in[0,1].
$$

Hence, $X$ represents the proportion of defective items in a batch.

---

## 3. Probability Density Function

The probability density function associated with $X$ is

$$
f_X(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x(1-x)^{17},
& 0<x<1,\\
0,
& \text{otherwise}.
\end{cases}
$$

---

## 4. Verification that $f$ is a Valid Density

To confirm that $f$ is a valid probability density function, observe first that

$$
f(x)\ge0
\qquad
\text{for all }x.
$$

Next, compute the total integral:

$$
\int_{-\infty}^{\infty} f(x)\,dx
=
\int_0^1
\dfrac{1}{B(2,18)}
x(1-x)^{17}\,dx.
$$

Using the definition of the Beta function,

$$
\int_{-\infty}^{\infty} f(x)\,dx
=
\dfrac{1}{B(2,18)}B(2,18)
=
1.
$$

Therefore, $f$ satisfies all requirements of a probability density function.

---

## 5. Expectation

Since

$$
X\sim\mathrm{Beta}(2,18),
$$

the mean of a Beta$(\alpha,\beta)$ distribution gives

$$
E[X]
=
\frac{\alpha}{\alpha+\beta}.
$$

Substituting $\alpha=2$ and $\beta=18$,

$$
E[X]
=
\frac{2}{20}
=
0.10.
$$

Thus,

$$
\boxed{E[X]=0.10.}
$$

---

## 6. Variance

For a Beta$(\alpha,\beta)$ random variable,

$$
\mathrm{Var}(X)
=
\frac{\alpha\beta}
{(\alpha+\beta)^2(\alpha+\beta+1)}.
$$

Hence,

$$
\mathrm{Var}(X)
=
\frac{2\cdot18}{20^2\cdot21}.
$$

Therefore,

$$
\mathrm{Var}(X)
=
\frac{36}{8400}
=
\frac{3}{700}
\approx0.00429.
$$

So,

$$
\boxed{\mathrm{Var}(X)\approx0.00429.}
$$

---

## 7. Probability that $X\le0.10$

The probability that the defective proportion does not exceed $0.10$ is

$$
P(X\le0.10)
=
\int_0^{0.10}
\dfrac{1}{B(2,18)}
x(1-x)^{17}\,dx.
$$

Equivalently,

$$
P(X\le0.10)
=
I_{0.10}(2,18),
$$

where $I_x(\alpha,\beta)$ denotes the regularized incomplete Beta function.

---

## 8. Probability that $0.05\le X\le0.15$

The probability that the defective proportion lies between $0.05$ and $0.15$ is

$$
P(0.05\le X\le0.15)
=
\int_{0.05}^{0.15}
\dfrac{1}{B(2,18)}
x(1-x)^{17}\,dx.
$$

Equivalently,

$$
P(0.05\le X\le0.15)
=
I_{0.15}(2,18)-I_{0.05}(2,18).
$$

---

## 9. Why the Beta Distribution is Appropriate

The Beta distribution is suitable in this situation because the defective proportion is a continuous variable restricted to the interval $[0,1]$.

Moreover, Beta distributions are commonly used for modeling proportions, probabilities, and rates.

---

## 10. Interpretation

The expected value

$$
E[X]=0.10
$$

indicates that the average defective proportion over the long run is $10\%$.

The variance

$$
\mathrm{Var}(X)\approx0.00429
$$

describes the variability of the defective proportion between batches, giving a measure of how consistent the manufacturing quality remains over time.

### The plot of the $\beta$-distribution

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import beta

def plot_beta_distribution(alpha, beta_val):
    """
    Plots the Beta distribution PDF for given alpha and beta parameters using Plotly.
    """
    # Create a range of x values from 0 to 1
    x = np.linspace(0, 1, 1000)

    # Calculate the PDF values for the Beta distribution
    y = beta.pdf(x, alpha, beta_val)

    # Create the Plotly figure
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=x,
        y=y,
        mode='lines',
        name=f'Beta($\\alpha={alpha}, \\beta={beta_val}$)',
        line=dict(width=2)
    ))

    # Update layout for better readability
    fig.update_layout(
        title=f'Beta Distribution Density Function ($\\alpha={alpha}, \\beta={beta_val}$)',
        xaxis_title='x',
        yaxis_title='PDF(x)',
        template='plotly_white'
    )

    # To display the plot, use fig.show()
    return fig

# Example usage:
fig = plot_beta_distribution(alpha=2, beta_val=18)
fig.show()
```

### Plot of the triangular distribution

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import triang

# Parameters given
a = 4
b = 10
c_mode = 6

# Scipy parametrization
# loc is the minimum (a)
# scale is the width (b - a)
# c_scipy is the ratio of the mode relative to the width (c - a) / (b - a)
loc = a
scale = b - a
c_scipy = (c_mode - a) / (b - a)

# Generate data points
x = np.linspace(0, b, 500)
y = triang.pdf(x, c=c_scipy, loc=loc, scale=scale)

# Create the Plotly figure
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode='lines',
    fill='tozeroy',
    name='Triangular Distribution',
    line=dict(color='royalblue', width=2)
))

# Layout updates
fig.update_layout(
    title=f'Triangular Distribution (a={a}, b={b}, c={c_mode})',
    xaxis_title='Time (hours)',
    yaxis_title='Probability Density',
    template='plotly_white'
)

# Display the plot
fig.show()
```

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


##  Answer

1. The parameter values are given by

$$
a=8,
\qquad
b=20,
\qquad
c=11.
$$

For the PERT distribution, the shape parameters are

$$
\alpha
=
1+4\left(\frac{c-a}{b-a}\right)
=
1+4\left(\frac{11-8}{20-8}\right).
$$

Hence,

$$
\alpha
=
1+4\left(\frac{3}{12}\right)
=
2.
$$

Similarly,

$$
\beta
=
1+4\left(\frac{b-c}{b-a}\right)
=
1+4\left(\frac{20-11}{20-8}\right).
$$

Therefore,

$$
\beta
=
1+4\left(\frac{9}{12}\right)
=
4.
$$

---

2. Since $X$ denotes the duration of the task in days, the possible values of $X$ satisfy

$$
8\le X\le20.
$$

---

3. The density function of the PERT model is

$$
f_X(x)
=
\frac{(x-8)^{\alpha-1}(20-x)^{\beta-1}}
{B(\alpha,\beta)(20-8)^{\alpha+\beta-1}},
\qquad
8\le x\le20.
$$

Substituting $\alpha=2$ and $\beta=4$ gives

$$
f_X(x)
=
\frac{(x-8)(20-x)^3}
{B(2,4)\,12^5},
\qquad
8\le x\le20.
$$

Now compute the Beta function:

$$
B(2,4)

Since

$$

\qquad
\Gamma(4)=3!,

$$

we obtain

$$
B(2,4)
=
\frac{1!\,3!}{5!}
=
\frac{6}{120}
=
\frac{1}{20}.
$$

Therefore,

$$
f_X(x)
=
\frac{20(x-8)(20-x)^3}{12^5},
\qquad

$$

and outside this interval,

$$
f_X(x)=0.
$$

---

4. A convenient probability space is

$$
\Omega=[8,20],
\qquad
\mathfrak F=\mathfrak B([8,20]),
$$

where $\mathfrak B([8,20])$ denotes the Borel $\sigma$-algebra on $[8,20]$.

Define the probability measure by

$$
P(A)=\int_A f_X(x)\,dx,
\qquad
A\in\mathfrak F.
$$

---

5. Define the random variable

$$
X:\Omega\to\mathbb R
$$

by

$$
X(\omega)=\omega,
\qquad
\omega\in[8,20].
$$

Thus, $X$ represents the completion time of the task in days.

---

6. The expected value is defined by

$$
E[X]
=
\int_a^b x f(x)\,dx.
$$

Substituting the density gives

$$
E[X]
=
\int_a^b
x
\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}
{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}
\,dx.
$$

To simplify the computation, introduce the transformation

$$
t=\frac{x-a}{b-a}.
$$

Then

$$
x=a+(b-a)t,
\qquad
dx=(b-a)\,dt.
$$

The interval transforms as follows:

- when $x=a$, we have $t=0$,
- when $x=b$, we have $t=1$.

Also,

$$
x-a=(b-a)t,
$$

and

$$
b-x=(b-a)(1-t).
$$

Substituting into the integral,

$$
E[X]
=
\int_0^1
\big(a+(b-a)t\big)
\frac{[(b-a)t]^{\alpha-1}[(b-a)(1-t)]^{\beta-1}}
{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}
(b-a)\,dt.
$$

Collecting the powers of $(b-a)$,

$$
E[X]
=
\frac{(b-a)^{\alpha-1+\beta-1+1}}
{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}
\int_0^1
\big(a+(b-a)t\big)
t^{\alpha-1}(1-t)^{\beta-1}\,dt.
$$

The factors cancel, leaving

$$
E[X]
=
\frac{1}{B(\alpha,\beta)}
\int_0^1
\left(
a\,t^{\alpha-1}(1-t)^{\beta-1}
+
(b-a)t^\alpha(1-t)^{\beta-1}
\right)\,dt.
$$

Using the Beta function identity,

$$
B(\alpha,\beta)
=
\int_0^1
t^{\alpha-1}(1-t)^{\beta-1}\,dt,
$$

we obtain

$$
E[X]
=
\frac{
aB(\alpha,\beta)
+
(b-a)B(\alpha+1,\beta)
}
{B(\alpha,\beta)}.
$$

Hence,

$$
E[X]
=
a+(b-a)\frac{B(\alpha+1,\beta)}{B(\alpha,\beta)}.
$$

Using

$$
B(x,y)
=
\frac{\Gamma(x)\Gamma(y)}{\Gamma(x+y)},
$$

it follows that

$$
\frac{B(\alpha+1,\beta)}{B(\alpha,\beta)}
=
\frac{\alpha}{\alpha+\beta}.
$$

Therefore,

$$
E[X]
=
a+(b-a)\frac{\alpha}{\alpha+\beta}.
$$

Now compute

$$
\alpha+\beta
=
\left(1+4\frac{c-a}{b-a}\right)
+
\left(1+4\frac{b-c}{b-a}\right).
$$

Simplifying,

$$
\alpha+\beta
=
2+4\left(\frac{c-a+b-c}{b-a}\right)
=
2+4(1)
=
6.
$$

Substituting into the expectation formula,

$$
E[X]
=
a+(b-a)\frac{1+4\frac{c-a}{b-a}}{6}.
$$

Expanding,

$$
E[X]
=
a+\frac{b-a}{6}+\frac{4(c-a)}{6}.
$$

Hence,

$$
E[X]
=
\frac{6a+b-a+4c-4a}{6}
=
\frac{a+4c+b}{6}.
$$

This is the standard PERT mean formula.

Substituting the numerical values,

$$
E[X]
=
\frac{8+4(11)+20}{6}
=
\frac{72}{6}
=
12.
$$

Thus,

$$
\boxed{E[X]=12.}
$$

---

7. The variance approximation for the PERT model is

$$
\mathrm{Var}(X)
=
\frac{(E[X]-a)(b-E[X])}{7}.
$$

Using $E[X]=12$, $a=8$, and $b=20$,

$$
\mathrm{Var}(X)
=
\frac{(12-8)(20-12)}{7}.
$$

Therefore,

$$
\mathrm{Var}(X)
=
\frac{4\cdot8}{7}
=
\frac{32}{7}
\approx4.57.
$$

---

8. The standard deviation is

$$
\sigma
=
\sqrt{\mathrm{Var}(X)}
=
\sqrt{\frac{32}{7}}
\approx2.14.
$$

Using the common approximation rule,

$$
\sigma
\approx
\frac{b-a}{6}
=
\frac{20-8}{6}
=
2,
$$

which is reasonably close to the exact value.

---

9.

| Quantity | Triangular | PERT |
|---|---:|---:|
| Mean | $13$ | $12$ |
| Variance | $\dfrac{49}{6}\approx8.17$ | $\dfrac{32}{7}\approx4.57$ |
| Standard deviation | $\approx2.86$ | $\approx2.14$ |
| Main feature | Linear with sharp peak | Smoother and more mode-focused |

The main observation is that the PERT distribution places greater emphasis on the most likely value, resulting in a smaller variance and a mean closer to the mode than the triangular model.

---

10. In this context:

- $a=8$ represents the minimum feasible completion time,
- $b=20$ represents the maximum feasible completion time,
- $c=11$ represents the most likely completion time,
- $E[X]=12$ gives the expected duration of the task,
- $\sigma\approx2.14$ measures the uncertainty in completion time.

These quantities are useful in project management because they help estimate realistic schedules, evaluate deadline risk, and determine whether additional contingency time is necessary.

### Plot of the triangle and PERT distributions

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import triang, beta

# Parameters
a = 4
b = 10
c_mode = 6

# --- Triangular Distribution Calculation ---
# Scipy parametrization for Triangular
c_scipy = (c_mode - a) / (b - a)
loc = a
scale = b - a

# --- PERT Distribution Calculation ---
# PERT is a Beta distribution with specific alpha and beta parameters
# alpha = 1 + 4 * (c - a) / (b - a)
# beta = 1 + 4 * (b - c) / (b - a)
alpha_pert = 1 + 4 * (c_mode - a) / (b - a)
beta_pert = 1 + 4 * (b - c_mode) / (b - a)

# Generate data points
x = np.linspace(0, b, 500)
y_triang = triang.pdf(x, c=c_scipy, loc=loc, scale=scale)
y_pert = beta.pdf(x, alpha_pert, beta_pert, loc=a, scale=b-a)

# Create the Plotly figure
fig = go.Figure()

# Add Triangular Distribution
fig.add_trace(go.Scatter(
    x=x,
    y=y_triang,
    mode='lines',
    name='Triangular Distribution',
    line=dict(color='royalblue', width=2, dash='dash')
))

# Add PERT Distribution
fig.add_trace(go.Scatter(
    x=x,
    y=y_pert,
    mode='lines',
    name='PERT Distribution',
    line=dict(color='firebrick', width=2)
))

# Layout updates
fig.update_layout(
    title=f'Comparison: Triangular vs PERT (a={a}, b={b}, c={c_mode})',
    xaxis_title='Time (hours)',
    yaxis_title='Probability Density',
    template='plotly_white'
)

# Display the plot
fig.show()
```

#Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

##  Answer

Start with the PERT expectation formula

$$
E[X]=\frac{a+4c+b}{6}.
$$

---

Suppose that, for the PERT distribution, the expected value is equal to the most likely value. Then

$$
E[X]=c.
$$

Substituting this into the mean formula gives

$$
\frac{a+4c+b}{6}=c.
$$

Multiplying both sides by $6$,

$$
a+4c+b=6c.
$$

Rearranging,

$$
a+b=2c.
$$

Hence,

$$
b=2c-a.
$$

Therefore,

$$
\boxed{b=17.}
$$

---

Now consider the variance approximation for the PERT model:

$$
\mathrm{Var}(X)
=
\frac{(E[X]-a)(b-E[X])}{7}.
$$

Since we impose the condition $E[X]=c$, this becomes

$$
\mathrm{Var}(X)
=
\frac{(c-a)(b-c)}{7}.
$$

Using the relation

$$
b=2c-a,
$$

we obtain

$$
b-c=c-a.
$$

Therefore,

$$
\mathrm{Var}(X)
=
\frac{(c-a)^2}{7}.
$$

An equivalent form is obtained by noting that

$$
b-a=2(c-a).
$$

Hence,

$$
(c-a)^2
=
\frac{(b-a)^2}{4}.
$$

Substituting into the variance expression,

$$
\mathrm{Var}(X)
=
\frac{(b-a)^2}{28}.
$$

---

Thus, when the expected value coincides with the most likely value, the PERT distribution becomes symmetric, and the variance simplifies to

$$
\boxed{
\mathrm{Var}_{\text{PERT}}(X)
=
\frac{(c-a)^2}{7}
}.
$$
