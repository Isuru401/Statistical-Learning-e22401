$$

$$

###Q1.5

To confirm that this random variable is indeed the conditional expectation of $X$ given $Y$, we check the defining condition

$$
\int_A E[X\mid Y]\,dP
=
\int_A X\,dP
\qquad \text{for every } A\in\sigma(Y).
$$

Since $Y$ takes only two values, the $\sigma$-algebra generated by $Y$ is

$$
\sigma(Y)
=
\left\{
\varnothing,\;
\Omega,\;
\{\omega_1,\omega_2\},\;
\{\omega_3,\omega_4\}
\right\}.
$$

---

For the empty set $A=\varnothing$,

$$
\int_{\varnothing} E[X\mid Y]\,dP
=
\int_{\varnothing} X\,dP
=
0.
$$

---

Next consider

$$
A=\{\omega_1,\omega_2\}.
$$

On this set,

$$
E[X\mid Y]=\frac16.
$$

Hence

$$
\int_A E[X\mid Y]\,dP
=
\frac16\,P(A)
=
\frac16(0.60)
=
0.10.
$$

On the other hand,

$$
\int_A X\,dP
=
X(\omega_1)P(\omega_1)

X(\omega_2)P(\omega_2).
$$

Since

$$
X(\omega_1)=0,
\qquad
X(\omega_2)=1,
$$

we obtain

$$
\int_A X\,dP
=
0(0.50)+1(0.10)
=
0.10.
$$

Thus the two integrals agree.

---

Now take

$$
A=\{\omega_3,\omega_4\}.
$$

Here,

$$
E[X\mid Y]=\frac12,
$$

so

$$
\int_A E[X\mid Y]\,dP
=
\frac12\,P(A)

\frac12(0.40)

0.20.

Also,

$$
\int_A X\,dP

Using

$$
X(\omega_3)=0,

X(\omega_4)=1,

we get

$$
\int_A X\,dP
=
0(0.20)+1(0.20)

0.20.
$$

Again the defining equality is satisfied.

---

Finally, for the whole sample space $A=\Omega$,

$$
\int_\Omega E[X\mid Y]\,dP
=
\frac16(0.60)+\frac12(0.40)

0.30.

Meanwhile,

$$
\int_\Omega X\,dP

0(0.50)+1(0.10)+0(0.20)+1(0.20)

Therefore,

$$
\int_A E[X\mid Y]\,dP

so the defining property of conditional expectation is verified.

As a consequence,

$$
E[E[X\mid Y]]

E[X],

which illustrates the Law of Total Expectation.

###Q1.8

We now compute the Minimum Mean Square Error (MMSE), which gives the smallest possible average squared estimation error when predicting $X$ using the conditional expectation $\mathbb{E}[X\mid Y]$.

The MMSE is

$$
\mathrm{MSE}_{\min}
=
\mathbb{E}\!\left[(X-\mathbb{E}[X\mid Y])^2\right]
=
\mathbb{E}[\mathrm{Var}(X\mid Y)].
$$

Thus we first determine the conditional variance for each value of $Y$.

Since $X$ is Bernoulli conditional on $Y=y$, its variance is

$$
\mathrm{Var}(X\mid Y=y)
=
p(1-p),
$$

where

$$
p=P(X=1\mid Y=y).
$$

---

For the low-risk class $Y=0$,

$$
P(X=1\mid Y=0)=\frac16.
$$

Therefore,

$$
\mathrm{Var}(X\mid Y=0)
=
\frac16\left(1-\frac16\right)
=
\frac16\cdot\frac56
=
\frac{5}{36}.
$$

---

For the high-risk class $Y=1$,

$$
P(X=1\mid Y=1)=\frac12.
$$

Hence,

$$
\mathrm{Var}(X\mid Y=1)
=
\frac12\left(1-\frac12\right)
=
\frac12\cdot\frac12
=
\frac14.
$$

---

Next we average these conditional variances using the probabilities of the corresponding risk classes.

Since

$$
P(Y=0)=0.60,
\qquad
P(Y=1)=0.40,
$$

we obtain

\begin{align*}
\mathbb{E}[\mathrm{Var}(X\mid Y)]
&=
\mathrm{Var}(X\mid Y=0)P(Y=0)
+
\mathrm{Var}(X\mid Y=1)P(Y=1)
\\[4pt]
&=
\frac{5}{36}(0.60)
+
\frac14(0.40)
\\[4pt]
&=
\frac{5}{36}\cdot\frac35
+
\frac14\cdot\frac25
\\[4pt]
&=
\frac{1}{12}
+
\frac{1}{10}
\\[4pt]
&=
\frac{5+6}{60}
\\[4pt]
&=
\frac{11}{60}
\approx 0.1833.
\end{align*}

Therefore,

$$
\mathrm{MSE}_{\min}
=
\frac{11}{60}
\approx 0.1833.
$$

This number represents the minimum achievable average squared prediction error. Any estimator based only on the risk class information $Y$ must have mean square error at least this large.

###Q1.9

Suppose the bank does not use the risk classification variable $Y$ and instead predicts fraud using the overall average

$$
\mathbb{E}[X]=0.30.
$$

In this case the estimator is the constant value

$$
\hat X = \mathbb{E}[X].
$$

The corresponding Mean Square Error becomes

$$
\mathrm{MSE}_{\text{global}}
=
\mathbb{E}\!\left[(X-\mathbb{E}[X])^2\right].
$$

Since the variance of a random variable is defined by

$$
\mathrm{Var}(X)
=
\mathbb{E}\!\left[(X-\mathbb{E}[X])^2\right],
$$

we obtain

$$
\mathrm{MSE}_{\text{global}}
=
\mathrm{Var}(X).
$$

---

Because $X$ is Bernoulli with

$$
P(X=1)=0.30,
$$

its variance is

$$
\mathrm{Var}(X)
=
p(1-p)
=
0.30(0.70)
=
0.21.
$$

Hence,

$$
\mathrm{MSE}_{\text{global}}=0.21.
$$

---

Previously we found that the optimal prediction error obtained using the information in $Y$ is

$$
\mathrm{MSE}_{\min}
=
\mathbb{E}[\mathrm{Var}(X\mid Y)]
=
\frac{11}{60}
\approx 0.1833.
$$

Therefore,

$$
0.1833 < 0.21,
$$

which shows that incorporating the risk class reduces the prediction error.

Thus the bank improves its estimation accuracy by using the additional information contained in $Y$.

---

This relationship is explained by the Law of Total Variance:

$$
\mathrm{Var}(X)
=
\mathbb{E}[\mathrm{Var}(X\mid Y)]
+
\mathrm{Var}(\mathbb{E}[X\mid Y]).
$$

The first term,

$$
\mathbb{E}[\mathrm{Var}(X\mid Y)],
$$

represents the uncertainty that remains after conditioning on $Y$. This is exactly the minimum achievable mean square error.

The second term,

$$
\mathrm{Var}(\mathbb{E}[X\mid Y]),
$$

measures how much variability is explained by the information in $Y$.

---

We now compute this second quantity.

Recall that

$$
\mathbb{E}[X\mid Y]
=
\begin{cases}
\dfrac16, & Y=0,\\[6pt]
\dfrac12, & Y=1,
\end{cases}
$$

with probabilities

$$
P(Y=0)=0.6,
\qquad
P(Y=1)=0.4.
$$

Hence,

\begin{align*}
\mathrm{Var}(\mathbb{E}[X\mid Y])
&=
\mathbb{E}\!\left[(\mathbb{E}[X\mid Y])^2\right]
-
(\mathbb{E}[X])^2
\\[6pt]
&=
\left(
\frac{1}{36}(0.6)
+
\frac14(0.4)
\right)
-
(0.3)^2
\\[6pt]
&=
(0.0167+0.10)-0.09
\\[6pt]
&\approx 0.0267.
\end{align*}

Therefore the total variance decomposition becomes

$$
0.21
=
0.1833
+
0.0267.
$$

---

This decomposition shows that the total prediction error can be separated into two parts:

\[
\text{Total Error}
=
\text{Remaining Error}
+
\text{Explained Error}.
\]

Thus the quantity

$$
\mathrm{Var}(\mathbb{E}[X\mid Y])
$$

measures the predictive value of the variable $Y$. The larger this term is, the more useful the risk classification becomes for predicting fraud.

### Q1.10 Quantifying Information Gain

To compute the information-theoretic quantities, we first organize the joint and marginal distributions of $X$ and $Y$.

---

## 1. Joint Distribution $P(X,Y)$

From the definitions of $X(\omega)$ and $Y(\omega)$, we directly obtain:

$$
P(X=0,Y=0)=P(\{\omega_1\})=0.50,
$$

$$
P(X=1,Y=0)=P(\{\omega_2\})=0.10,
$$

$$
P(X=0,Y=1)=P(\{\omega_3\})=0.20,
$$

$$
P(X=1,Y=1)=P(\{\omega_4\})=0.20.
$$

---

## Marginal Distributions

By summation over the joint probabilities:

### For $Y$:
$$
P(Y=0)=0.60,
\qquad
P(Y=1)=0.40.
$$

### For $X$:
$$
P(X=0)=0.70,
\qquad
P(X=1)=0.30.
$$

---

## 2. Conditional Entropy $H(X\mid Y)$

Conditional entropy is computed as the weighted average of entropies conditioned on each value of $Y$.

---

### Case $Y=0$

We have

$$
P(X=1\mid Y=0)=\frac16,
\qquad
P(X=0\mid Y=0)=\frac56.
$$

Thus,

$$
H(X\mid Y=0)
=
-\left(
\frac16\log_2\frac16
+
\frac56\log_2\frac56
\right)
\approx 0.650 \text{ bits}.
$$

---

### Case $Y=1$

We have

$$
P(X=1\mid Y=1)=\frac12,
\qquad
P(X=0\mid Y=1)=\frac12.
$$

Hence,

$$
H(X\mid Y=1)
=
-\left(
\frac12\log_2\frac12
+
\frac12\log_2\frac12
\right)
=
1 \text{ bit}.
$$

---

### Combine the Results

\begin{align*}
H(X\mid Y)
&=
P(Y=0)H(X\mid Y=0)
+
P(Y=1)H(X\mid Y=1)
\\[6pt]
&=
0.60(0.650)
+
0.40(1)
\\[6pt]
&=
0.390
+
0.400
\\[6pt]
&=
0.790 \text{ bits}.
\end{align*}

---

## 3. Mutual Information $I(X;Y)$

Mutual information is defined as

$$
I(X;Y)=H(X)-H(X\mid Y).
$$

---

### Step A: Entropy of $X$

$$
H(X)
=
-(0.70\log_2 0.70 + 0.30\log_2 0.30)
\approx 0.881 \text{ bits}.
$$

---

### Step B: Compute Mutual Information

$$
I(X;Y)
=
0.881 - 0.790
=
0.091 \text{ bits}.
$$

---

## 4. Interpretation

- Total uncertainty:
$$
H(X)=0.881
$$

- Remaining uncertainty after observing $Y$:
$$
H(X\mid Y)=0.790
$$

- Information gained from $Y$:
$$
I(X;Y)=0.091
$$

Thus, observing $Y$ reduces uncertainty in $X$ by

$$
\frac{0.091}{0.881}\approx 10.3\%.
$$

---

## Conclusion

The positive mutual information

$$
I(X;Y)>0
$$

confirms that the risk class $Y$ carries useful predictive information about fraud $X$, making it a relevant feature in the bank’s detection model.

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

## 2.1 Marginal Density of $X$

Let
$$
\widetilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_x,
\qquad
\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_y.
$$

Using the block structure of the covariance matrix, the quadratic form $Q$ can be separated into two components:

$$
Q
=
\widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}}
+
(\widetilde{\mathbf{y}} - \mathbf{A}\widetilde{\mathbf{x}})^T
\mathbf{S}_{YY}^{-1}
(\widetilde{\mathbf{y}} - \mathbf{A}\widetilde{\mathbf{x}}),
$$

where

$$
\mathbf{A} = \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}.
$$

---

With this decomposition, the joint density becomes

$$
f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})
=
C
\exp\!\left(
-\frac{1}{2}
\widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}}
\right)
\exp\!\left(
-\frac{1}{2}
(\widetilde{\mathbf{y}}-\mathbf{m})^T
\mathbf{S}_{YY}^{-1}
(\widetilde{\mathbf{y}}-\mathbf{m})
\right),
$$

where

$$
\mathbf{m} = \mathbf{A}\widetilde{\mathbf{x}}.
$$

---

## Marginalization over $\mathbf{Y}$

The marginal density of $\mathbf{X}$ is defined by

$$
f_{\mathbf{X}}(\mathbf{x})
=
\int_{\mathbb{R}^m}
f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})
\, d\mathbf{y}.
$$

Substituting the factorized form, we obtain

$$
f_{\mathbf{X}}(\mathbf{x})
=
C
\exp\!\left(
-\frac{1}{2}
\widetilde{\mathbf{x}}^T \boldsymbol{\Sigma}_{XX}^{-1} \widetilde{\mathbf{x}}
\right)
\int_{\mathbb{R}^m}
\exp\!\left(
-\frac{1}{2}
(\widetilde{\mathbf{y}}-\mathbf{m})^T
\mathbf{S}_{YY}^{-1}
(\widetilde{\mathbf{y}}-\mathbf{m})
\right)
d\mathbf{y}.
$$

---

The integral is the normalization constant of a multivariate normal density in $\mathbf{y}$ with covariance $\mathbf{S}_{YY}$, hence

$$
\int_{\mathbb{R}^m} (\cdots)\, d\mathbf{y}
=
\sqrt{(2\pi)^m |\mathbf{S}_{YY}|}.
$$

---

Using the block determinant identity

$$
|\boldsymbol{\Sigma}|
=
|\boldsymbol{\Sigma}_{XX}|\cdot|\mathbf{S}_{YY}|,
$$

we combine constants to get

$$
f_{\mathbf{X}}(\mathbf{x})
=
\frac{\sqrt{(2\pi)^m |\mathbf{S}_{YY}|}}
{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}_{XX}| |\mathbf{S}_{YY}|}}
\exp\!\left(
-\frac{1}{2}
(\mathbf{x}-\boldsymbol{\mu}_x)^T
\boldsymbol{\Sigma}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}_x)
\right).
$$

Simplifying yields

$$
f_{\mathbf{X}}(\mathbf{x})
=
\frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{XX}|}}
\exp\!\left(
-\frac{1}{2}
(\mathbf{x}-\boldsymbol{\mu}_x)^T
\boldsymbol{\Sigma}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}_x)
\right).
$$

---

## Conclusion

Therefore, the marginal distribution of $\mathbf{X}$ is

$$
\mathbf{X} \sim \mathcal{N}(\boldsymbol{\mu}_x,\boldsymbol{\Sigma}_{XX}).
$$

This shows that any subvector of a jointly Gaussian vector is itself Gaussian, with mean and covariance given by the corresponding blocks of the full covariance matrix.

## 2.2 Marginal Density of $Y$

The derivation for $f_{\mathbf{Y}}(\mathbf{y})$ is mathematically symmetric to the derivation for $\mathbf{X}$. The goal is to integrate out $\mathbf{x}$ to isolate the distribution of $\mathbf{Y}$.


We use the **Schur complement of $\boldsymbol{\Sigma}_{YY}$** (instead of $\boldsymbol{\Sigma}_{XX}$) to factor the quadratic form $Q$. The inverse covariance matrix $\boldsymbol{\Sigma}^{-1}$ is decomposed such that the terms involving $\mathbf{x}$ are "completed" into a square:

$$Q = \widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}} + (\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})^T \mathbf{S}_{XX}^{-1} (\widetilde{\mathbf{x}} - \mathbf{B}\widetilde{\mathbf{y}})$$

Where:
* $\widetilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_y$
* $\mathbf{B} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$ (the regression coefficients of $\mathbf{X}$ on $\mathbf{Y}$)
* $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$ (the Schur complement of $\boldsymbol{\Sigma}_{YY}$)

---

The marginal $f_{\mathbf{Y}}(\mathbf{y})$ is calculated by integrating the joint density over $d\mathbf{x}$:

$$f_{\mathbf{Y}}(\mathbf{y}) = \int_{\mathbb{R}^n} f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y}) d\mathbf{x}$$

The term $\exp\left(-\frac{1}{2} \widetilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \widetilde{\mathbf{y}}\right)$ is constant with respect to $\mathbf{x}$ and moves outside the integral. The remaining integral over $\mathbb{R}^n$ covers the kernel of a Gaussian in $\mathbf{X}$, which evaluates to $\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}$.


Using the determinant identity $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$, the constants cancel out just as before, leaving:

$$f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y) \right)$$



# 3. Conditional Density

The conditional density of $\mathbf{X}$ given $\mathbf{Y}=\mathbf{y}$ is defined by

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}\mid \mathbf{y})
=
\frac{f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})}
{f_{\mathbf{Y}}(\mathbf{y})}.
$$

Thus, deriving the conditional distribution reduces to combining the previously obtained joint and marginal Gaussian forms.

---

## Step 1: Structure of the joint and marginal exponents

From the Schur complement decomposition, the joint quadratic form can be written as

$$
Q_{\text{joint}}
=
(\mathbf{y}-\boldsymbol{\mu}_y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y)
+
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\mathbf{S}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*),
$$

where

$$
\mathbf{S}_{XX}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX},
$$

and

$$
\boldsymbol{\mu}^*
=
\boldsymbol{\mu}_x
+
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y).
$$

The marginal exponent of $\mathbf{Y}$ is

$$
Q_{\text{marginal}}
=
(\mathbf{y}-\boldsymbol{\mu}_y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y).
$$

---

## Step 2: Cancellation in the exponent

Using the identity

$$
\frac{e^{A}}{e^{B}} = e^{A-B},
$$

the conditional exponent becomes

\begin{align*}
Q_{\text{conditional}}
&=
Q_{\text{joint}} - Q_{\text{marginal}}
\\[6pt]
&=
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\mathbf{S}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*).
\end{align*}

Thus, all terms involving only $\mathbf{y}$ cancel out.

---

## Step 3: Normalization constants

The joint density constant is

$$
\frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}},
$$

and the marginal density constant is

$$
\frac{1}{\sqrt{(2\pi)^m|\boldsymbol{\Sigma}_{YY}|}}.
$$

Dividing gives

\begin{align*}
\frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}}
\cdot
\sqrt{(2\pi)^m|\boldsymbol{\Sigma}_{YY}|}
&=
\frac{1}{\sqrt{(2\pi)^n \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}.
\end{align*}

Using the determinant identity

$$
|\boldsymbol{\Sigma}|
=
|\boldsymbol{\Sigma}_{YY}|\cdot|\mathbf{S}_{XX}|,
$$

we obtain

$$
\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}
=
|\mathbf{S}_{XX}|.
$$

---

## Step 4: Final conditional density

Substituting both the simplified exponent and constant yields

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}\mid \mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^n|\mathbf{S}_{XX}|}}
\exp\!\left(
-\frac{1}{2}
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\mathbf{S}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*)
\right).
$$

Therefore,

$$
(\mathbf{X}\mid \mathbf{Y}=\mathbf{y})
\sim
\mathcal{N}
\!\left(
\boldsymbol{\mu}_x +
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y),
\;
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right).
$$

---

## Summary of parameters

### Conditional mean
$$
\boldsymbol{\mu}_{X|Y}
=
\boldsymbol{\mu}_x
+
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y).
$$

### Conditional covariance
$$
\boldsymbol{\Sigma}_{X|Y}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

A key observation is that $\boldsymbol{\Sigma}_{X|Y}$ does not depend on the specific value of $\mathbf{y}$. It depends only on the joint covariance structure.

This invariance is central in linear regression and Kalman filtering: observing $\mathbf{y}$ shifts the mean estimate, but the uncertainty structure (variance reduction) remains fixed for a given model.

# 4. Conditional Expectation

Let
$$
g(y)=\mathbb{E}[X\mid Y=y].
$$

By the Disintegration Theorem, the conditional expectation can be written as an integral with respect to the conditional density:

$$
g(y)
=
\int_{\mathbb{R}^n}
\mathbf{x}\,
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}\mid \mathbf{y})
\, d\mathbf{x}.
$$

---

Since the conditional distribution is Gaussian,

$$
(\mathbf{X}\mid \mathbf{Y}=\mathbf{y})
\sim
\mathcal{N}
\!\left(
\boldsymbol{\mu}_x +
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y),
\;
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right),
$$

the integral above reduces to the mean of this Gaussian distribution.

---

## Step 1: Use the Gaussian mean property

For any multivariate normal random vector, the expectation equals its mean parameter. Hence,

$$
g(y)
=
\boldsymbol{\mu}_x +
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y).
$$

---

## Step 2: Conditional expectation as a function

Therefore,

$$
\mathbb{E}[X\mid Y=y]
=
\boldsymbol{\mu}_x +
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_y).
$$

---

## Step 3: Random-variable form

Replacing $y$ by the random variable $Y$, we obtain the conditional expectation as a function of $Y$:

$$
\mathbb{E}[X\mid Y]
=
\boldsymbol{\mu}_x +
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{Y}-\boldsymbol{\mu}_y).
$$

---

## Conclusion

The conditional expectation is therefore an affine (linear + shift) function of $Y$:

$$
\mathbb{E}[X\mid Y]
=
\boldsymbol{\mu}_x +
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{Y}-\boldsymbol{\mu}_y),
$$

which shows that in the Gaussian case, the optimal estimator is linear in the observed variable.

# 5. The Conditional Variances

Since
$$
(\mathbf{X}\mid \mathbf{Y}=\mathbf{y})
\sim
\mathcal{N}
\!\left(
\boldsymbol{\mu}_x +
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y}-\boldsymbol{\mu}_y),
\;
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
\right),
$$

we immediately obtain the conditional covariance:

$$
\mathrm{Var}(\mathbf{X}\mid \mathbf{Y}=\mathbf{y})
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

---

A key observation is that this expression does not depend on $\mathbf{y}$. Hence,

$$
\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

Taking expectation preserves the same value:

$$
\mathbb{E}\!\left[\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})\right]
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

---

## Estimation Error

The mean squared estimation error of the optimal predictor is

\begin{align*}
\mathbb{E}\big[(\mathbf{X}-\mathbb{E}[\mathbf{X}\mid \mathbf{Y}])^2\big]
&=
\mathbb{E}\!\left[\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})\right]
\\[6pt]
&=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
\end{align*}

---

## Law of Total Variance

For any random vectors with finite second moments, the variance decomposes as

$$
\mathrm{Var}(\mathbf{X})
=
\mathbb{E}[\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})]
+
\mathrm{Var}(\mathbb{E}[\mathbf{X}\mid \mathbf{Y}]).
$$

Rearranging gives

$$
\mathrm{Var}(\mathbb{E}[\mathbf{X}\mid \mathbf{Y}])
=
\mathrm{Var}(\mathbf{X})
-
\mathbb{E}[\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})].
$$

Substituting the Gaussian expressions yields

$$
\mathrm{Var}(\mathbb{E}[\mathbf{X}\mid \mathbf{Y}])
=
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.
$$

---

## Final Decomposition

We therefore obtain the clean split:

### Explained variance
$$
\mathrm{Var}(\mathbb{E}[\mathbf{X}\mid \mathbf{Y}])
=
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.
$$

### Unexplained variance (estimation error)
$$
\mathbb{E}[\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})]
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.
$$

---

## Interpretation

The total uncertainty in $\mathbf{X}$ is decomposed into:

- **Explained part:** variability captured by the linear dependence on $\mathbf{Y}$
- **Unexplained part:** residual uncertainty remaining after observing $\mathbf{Y}$

Thus the matrix
$$
\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}
$$
quantifies exactly how much of the variability in $\mathbf{X}$ can be explained through $\mathbf{Y}$ in the Gaussian setting.

# 6. Entropt of the Conditional Expectation and Mutual Information

Let
$$
\mathbf{Z}
=
\begin{bmatrix}
\mathbf{X}\\
\mathbf{Y}
\end{bmatrix}
\in \mathbb{R}^{n+m},
\qquad
\mathbf{Z} \sim \mathcal{N}(\boldsymbol{\mu},\boldsymbol{\Sigma}).
$$

The joint density is given by

$$
f_{\mathbf{X},\mathbf{Y}}(\mathbf{z})
=
\frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}}
\exp\!\left(
-\frac{1}{2}
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu})
\right).
$$

---

## Differential Entropy Definition

The differential entropy is defined as

$$
h(\mathbf{Z})
=
-\mathbb{E}[\ln f_{\mathbf{X},\mathbf{Y}}(\mathbf{Z})].
$$

---

## Step 1: Log-density

Taking the natural logarithm of the density gives

$$
\ln f_{\mathbf{X},\mathbf{Y}}(\mathbf{z})
=
-\frac{1}{2}\ln\!\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
-\frac{1}{2}
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu}).
$$

---

## Step 2: Expectation

Substituting into the definition of entropy:

\begin{align*}
h(\mathbf{Z})
&=
-\mathbb{E}[\ln f_{\mathbf{X},\mathbf{Y}}(\mathbf{Z})]
\\[6pt]
&=
\mathbb{E}\left[
\frac{1}{2}\ln\!\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
+
\frac{1}{2}
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
\right].
\end{align*}

Since the logarithmic term is constant, it can be taken outside the expectation:

\begin{align*}
h(\mathbf{Z})
&=
\frac{1}{2}\ln\!\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
+
\frac{1}{2}
\mathbb{E}\!\left[
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
\right].
\end{align*}

---

## Step 3: Quadratic form expectation

For a multivariate Gaussian random vector,

$$
\mathbb{E}\!\left[
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
\right]
=
\mathrm{Tr}(\boldsymbol{\Sigma}^{-1}\boldsymbol{\Sigma})
=
\mathrm{Tr}(\mathbf{I}_{n+m})
=
n+m.
$$

---

## Step 4: Final expression

Substituting this result gives

\begin{align*}
h(\mathbf{Z})
&=
\frac{1}{2}\ln\!\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
+
\frac{n+m}{2}.
\end{align*}

Writing $\frac{n+m}{2} = \frac{1}{2}\ln(e^{n+m})$, we combine terms:

$$
h(\mathbf{Z})
=
\frac{1}{2}
\ln\!\left((2\pi e)^{n+m}|\boldsymbol{\Sigma}|\right).
$$

---

## Conclusion

Thus, the differential entropy of a multivariate normal random vector is

$$
h(\mathbf{Z})
=
\frac{1}{2}
\ln\!\left((2\pi e)^{n+m}|\boldsymbol{\Sigma}|\right),
$$

showing that entropy depends only on the covariance structure through $|\boldsymbol{\Sigma}|$ and increases with the volume of uncertainty in the distribution.

## 6.2 Conditional Entropy

The conditional entropy $h(\mathbf{X}\mid\mathbf{Y})$ represents the average amount of uncertainty remaining in $\mathbf{X}$ after $\mathbf{Y}$ is known. For a multivariate Gaussian, this is determined by the varaince of the contional distribution
$$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathscr{N}(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX})$$

The conditional covariance $\Sigma_{X|Y}$ covariance is thus given by $$\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.$$  
Since $\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}$ is independent of the specific value $y$, the uncertainty is constant across all observations of $Y$.  The differential entropy for this conditional distribution is thus :
$$h(\mathbf{X}\mid\mathbf{Y}) = \frac{1}{2} \ln((2\pi e)^n |\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}).$$

## 6.3 The Entropy of the Conditional Expectation

From the law of total variance we have seen that the explained varaince is:
$\text{Var}(\mathbb{E}[\mathbf{X}\mid\mathbf{Y}]) = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$. Thus the differential entropy of $\mathbb{E}(\mathbf{X}\mid\mathbf{Y})$is:
$$h(\mathbb{E}(\mathbf{X}\mid\mathbf{Y})) = \frac{1}{2} \ln((2\pi e)^n |\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|).$$

## 6.4 Mutual Information

Mutual information is given by:

$$I(\mathbf{X};\mathbf{Y}) = h(\mathbf{X}) - h(\mathbf{X} \mid \mathbf{Y}).$$

Substitute the entropy formulas for the marginal and conditional distributions derived in the document:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{XX}|\right) - \frac{1}{2} \ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}|\right) \text{.}$$

Using the property of logarithms ($\ln A - \ln B = \ln \frac{A}{B}$), the constants $(2\pi e)^n$ cancel out:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln \left( \frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}|} \right) \text{.}$$

Substituting the expression for the conditional covariance $\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln \left( \frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|} \right).$$

\mathbb{E}[X^2] + \mathbb{E}[XQ].
\end{align*}

Since $\mathbb{E}[X^2]=1$ and independence gives $\mathbb{E}[XQ]=0$, we obtain
$$
\mathrm{Cov}(X,Y)=1.
$$

---

Thus the joint distribution is
$$
\begin{pmatrix}
X\\
Y
\end{pmatrix}
\sim
\mathcal{N}
\!\left(
\begin{pmatrix}
0\\
0
\end{pmatrix},
\begin{pmatrix}
1 & 1\\
1 & 5
\end{pmatrix}
\right).
$$

---

\textbf{2. Information Structure}

The variable $X$ is not $\sigma(Y)$-measurable because $Y=X+Q$ does not uniquely determine $X$. For any observed $Y=y$, infinitely many decompositions $(x,q)$ satisfy $x+q=y$.

Thus, $Y$ provides partial but not complete information about $X$.

---

\textbf{3. Conditional Expectation}

For jointly Gaussian variables,
$$
\mathbb{E}[X\mid Y=y]
=
\mu_X
+
\frac{\mathrm{Cov}(X,Y)}{\mathrm{Var}(Y)}
(y-\mu_Y).
$$

Substituting values gives
\begin{align*}
\mathbb{E}[X\mid Y=y]
&=
0 + \frac{1}{5}(y-0)
\\[6pt]
&=
0.2y.
\end{align*}

Hence,
$$
\mathbb{E}[X\mid Y] = 0.2Y.
$$

---

\textbf{4. Residual Verification}

Let $\widehat{X}=0.2Y$ and define $Z=X-\widehat{X}$.

---

\textbf{Mean of residual:}
$$
\mathbb{E}[Z]
=
\mathbb{E}[X] - 0.2\mathbb{E}[Y]
=
0.
$$

---

\textbf{Orthogonality check:}
\begin{align*}
\mathrm{Cov}(Z,Y)
&=
\mathbb{E}[(X-0.2Y)Y]
\\[6pt]
&=
\mathbb{E}[XY] - 0.2\mathbb{E}[Y^2].
\end{align*}

Using $\mathbb{E}[XY]=1$ and $\mathbb{E}[Y^2]=5$, we get
$$
\mathrm{Cov}(Z,Y)=1-0.2(5)=0.
$$

Thus the estimation error is orthogonal to $Y$.

---

\textbf{5. Differential Entropy}

For a Gaussian variable,
$$
h(\mathcal{N}(0,\sigma^2))
=
\frac{1}{2}\ln(2\pi e\sigma^2).
$$

---

\textbf{Marginal entropy of $X$:}
$$
h(X)
=
\frac{1}{2}\ln(2\pi e)
\approx 1.4189.
$$

---

\textbf{Conditional variance:}
$$
\sigma_{X|Y}^2
=
1 - \frac{1^2}{5}
=
0.8.
$$

---

\textbf{Conditional entropy:}
$$
h(X|Y)
=
\frac{1}{2}\ln(2\pi e \cdot 0.8)
\approx 1.3073.
$$

---

\textbf{6. Mutual Information}

\begin{align*}
I(X;Y)
&=
h(X) - h(X|Y)
\\[6pt]
&=
\frac{1}{2}\ln\left(\frac{1}{0.8}\right)
\\[6pt]
&=
\frac{1}{2}\ln(1.25)
\\[6pt]
&\approx 0.1116 \text{ nats}.
\end{align*}

---

\textbf{7. Interpretation}

- Using $X=Y$ ignores noise and gives high error:
$$
\mathbb{E}[(X-Y)^2] = \mathrm{Var}(Q)=4.
$$

- Using the optimal estimator gives:
$$
\mathrm{Var}(X-\mathbb{E}[X|Y]) = 0.8.
$$

Thus, the optimal linear estimator significantly reduces uncertainty by filtering out noise from the observation $Y$.

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal, norm

# --- 1. Parameters for the Spindle Problem ---
# X: Temperature ~ N(0, 1)
# Q: Torque ~ N(0, 4)
# Y: Power = X + Q => Y ~ N(0, 5)
mu_x, mu_y = 0, 0
var_x, var_q = 1, 4
var_y = var_x + var_q
cov_xy = var_x  # Cov(X, X+Q) = Var(X) = 1
rho = cov_xy / (np.sqrt(var_x) * np.sqrt(var_y))

# Shrinkage Factor (Gain)
k = cov_xy / var_y  # 1/5 = 0.2

# --- 2. Generate Data for Visualization ---
# Sample points to show the "Cloud" of data
np.random.seed(42)
n_samples = 1000
x_samples = np.random.normal(mu_x, np.sqrt(var_x), n_samples)
q_samples = np.random.normal(0, np.sqrt(var_q), n_samples)
y_samples = x_samples + q_samples
x_hat = k * y_samples  # E[X|Y]
residual = x_samples - x_hat

# Grids for Contours
x_grid = np.linspace(-4, 4, 200)
y_grid = np.linspace(-8, 8, 200)
X_m, Y_m = np.meshgrid(x_grid, y_grid)
pos = np.dstack((X_m, Y_m))
Z = multivariate_normal([mu_x, mu_y], [[var_x, cov_xy], [cov_xy, var_y]]).pdf(pos)

# --- 3. Create Simulation Plot ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Joint Distribution: X vs Y<br><sup>The Line of Best Estimate (Shrinkage)</sup>",
        "Estimation Error (Residual) vs Y<br><sup>Visualizing the Orthogonality Principle</sup>"
    )
)

# --- LEFT PLOT: Joint X, Y and the Estimator ---
# Density Contour
fig.add_trace(go.Contour(
    x=x_grid, y=y_grid, z=Z, colorscale='Blues', showscale=False, opacity=0.6,
    hoverinfo='skip'
), row=1, col=1)

# Scatter of actual realizations
fig.add_trace(go.Scatter(
    x=x_samples, y=y_samples, mode='markers',
    marker=dict(color='rgba(0,0,0,0.1)', size=3),
    name="Realized (x, y)"
), row=1, col=1)

# The Best Estimate Line E[X|Y] = 0.2Y
y_line = np.linspace(-8, 8, 10)
fig.add_trace(go.Scatter(
    x=k * y_line, y=y_line,
    line=dict(color='Red', width=4),
    name="E[X|Y] = 0.2Y (Optimal)"
), row=1, col=1)

# The Naive Estimator Line X = Y
fig.add_trace(go.Scatter(
    x=y_line, y=y_line,
    line=dict(color='Black', width=2, dash='dash'),
    name="X = Y (Naive)"
), row=1, col=1)

# --- RIGHT PLOT: Orthogonality (Residual vs Y) ---
fig.add_trace(go.Scatter(
    x=y_samples, y=residual, mode='markers',
    marker=dict(color='Purple', opacity=0.4, size=4),
    name="Residual (X - 0.2Y)"
), row=1, col=2)

# Horizontal line at 0 to show no correlation
fig.add_trace(go.Scatter(
    x=y_line, y=np.zeros_like(y_line),
    line=dict(color='Black', width=2),
    name="Zero Mean Error"
), row=1, col=2)

# --- Layout and Styling ---
fig.update_layout(
    height=600,
    title_text="Spindle Motor Diagnostic Simulation: Extracting Signal from Noise",
    template="plotly_white",
    hovermode='x'
)

fig.update_xaxes(title_text="True Temperature (X)", row=1, col=1)
fig.update_yaxes(title_text="Observed Power (Y)", row=1, col=1)
fig.update_xaxes(title_text="Observed Power (Y)", row=1, col=2)
fig.update_yaxes(title_text="Residual Error (Z)", row=1, col=2)

fig.show()
```

The simulation results provide a clear connection between probabilistic estimation theory and its physical interpretation in noisy industrial systems.

## 1. Shrinkage Principle (Conservative Estimation)

We compare the naive estimator and the optimal estimator:

$$
\text{Naive: } \widehat{X} = Y,
\qquad
\text{Optimal: } \mathbb{E}[X\mid Y] = 0.2Y.
$$

Since the noise variance dominates the signal,

$$
\mathrm{Var}(Q)=4 \gg \mathrm{Var}(X)=1,
$$

the observation $Y=X+Q$ is heavily contaminated.

Thus the optimal estimator applies a shrinkage factor:

$$
\widehat{X} = 0.2Y.
$$

---

### Interpretation

A large observed value (e.g. $Y=5$) is not fully attributed to $X$. Instead,

$$
\mathbb{E}[X\mid Y=5] = 1,
$$

meaning most of the observed spike is attributed to noise $Q$, not signal $X$.

This mechanism reduces false alarms by preventing overreaction to noisy fluctuations.

---

## 2. Orthogonality Principle (Optimality Condition)

Let the estimation error be

$$
Z = X - \mathbb{E}[X\mid Y].
$$

Then

$$
\mathrm{Cov}(Z,Y)=0.
$$

---

### Interpretation

This implies that no linear function of $Y$ can further reduce the error. The residual $Z$ contains only information that is independent of $Y$, meaning it behaves like pure noise.

Therefore, the estimator achieves the minimum mean square error (MMSE).

---

## 3. Information Geometry and Loss of Invertibility

The mapping

$$
(X,Q) \mapsto Y = X + Q
$$

reduces a two-dimensional system to one dimension, implying a loss of information.

Even when $Y$ is fixed, $X$ is still random:

$$
\mathrm{Var}(X\mid Y)=0.8.
$$

---

### Interpretation

The conditional distribution represents a vertical "slice" of uncertainty. The width of this slice corresponds to irreducible uncertainty, which cannot be removed by any estimator based solely on $Y$.

---

## 4. Variance Reduction and Information Gain

Before observing $Y$:

$$
\mathrm{Var}(X)=1.
$$

After observing $Y$:

$$
\mathrm{Var}(X\mid Y)=0.8.
$$

Thus the reduction in uncertainty is

$$
I(X;Y)=0.1116 \text{ nats}.
$$

---

### Interpretation

Only 20% of the original uncertainty is removed by measuring $Y$. The remaining uncertainty is intrinsic noise that cannot be eliminated without introducing a different measurement mechanism.
