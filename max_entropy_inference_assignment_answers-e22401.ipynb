# Q. Derivation of the MaxEnt Solution

We seek a discrete probability distribution $p_i = P(X = x_i)$ for $i = 1, \dots, k$ that maximizes the **Entropy** $H(P)$:

$$\max_{p_1, \dots, p_k} H(P) = -\sum_{i=1}^{k} p_i \ln p_i$$

Subject to the normalization constraint and $m$ macrostate measurement constraints:

$$\sum_{i=1}^{k} p_i = 1 \quad \text{and} \quad \mathbb{E}[{f}_j(X)]=\sum_{i=1}^{k} p_i f_j(x_i) = \langle f_j \rangle \quad (j = 1, \dots, m).$$


**Show the Following:**


1. The **MaxEnt probability distribution** is:
$$p_j^* = \frac{1}{Z} \exp\left(-\sum_{i=1}^m \lambda_i f_i(x_j)\right),$$
where the **Partition Function**, $Z$:


$$Z(\lambda_1, \dots, \lambda_m) \triangleq \sum_{j=1}^k \exp\left(-\sum_{i=1}^m \lambda_i f_i(x_j)\right)$$


2. **Expectation Values:**

$$\frac{\partial \ln Z}{\partial \lambda_i} =  -\langle f_i(x)\rangle$$


3. **Fluctuations (Variance):**

$$\frac{\partial^2 \ln Z}{\partial \lambda_i^2} = \langle f_i(x)\rangle^2 - \langle f_i(x)^2\rangle = \operatorname{Var}(f_i(x))$$


4. **Maximized Thermodynamic Entropy:**

$$H_{max} = \ln Z + \sum_{i=1}^m \lambda_i \langle f_i(x)\rangle$$



5. If the microstate properties $f_i(x_j, \alpha_1, \dots, \alpha_r)$ depend smoothly on external parameters $\alpha$, the total differential of the entropy expands cleanly to:


$$dH = \sum_{i=1}^m \lambda_i d\langle f_i(x)\rangle + \sum_{r} \frac{\partial \ln Z}{\partial \alpha_r} d\alpha_r$$

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

# Maximum Entropy Distribution with Fixed Mean and Variance

Let $p(x)$ be the probability density function of the continuous random variable $X$.

The differential entropy is

$$
H(X)=-\int_{-\infty}^{\infty}p(x)\ln p(x),dx.
$$

We want to maximize $H(X)$ subject to the following constraints:

### Normalization condition

$$
\int_{-\infty}^{\infty}p(x),dx=1.
$$

### Mean condition

$$
\int_{-\infty}^{\infty}x,p(x),dx=\mu.
$$

Equivalently,

$$
\int_{-\infty}^{\infty}(x-\mu)p(x),dx=0.
$$

### Variance condition

$$
\int_{-\infty}^{\infty}(x-\mu)^2p(x),dx=\sigma^2.
$$

---

## Step 1: Form the Lagrange Functional

Introduce the Lagrange multipliers $\lambda_0$, $\lambda_1$, and $\lambda_2$.

$$
\begin{aligned}
J[p]
=&-\int_{-\infty}^{\infty}p(x)\ln p(x),dx\
&+\lambda_0
\left[
\int_{-\infty}^{\infty}p(x),dx-1
\right]\
&+\lambda_1
\int_{-\infty}^{\infty}(x-\mu)p(x),dx\
&+\lambda_2
\left[
\int_{-\infty}^{\infty}(x-\mu)^2p(x),dx-\sigma^2
\right].
\end{aligned}
$$

---

## Step 2: Differentiate with Respect to $p(x)$

For maximum entropy, the functional derivative must satisfy

$$
\frac{\delta J}{\delta p(x)}=0.
$$

Therefore,

$$
-\left(\ln p(x)+1\right)
+\lambda_0
+\lambda_1(x-\mu)
+\lambda_2(x-\mu)^2
=0.
$$

Rearranging,

$$
\ln p(x)
========

\lambda_0-1
+\lambda_1(x-\mu)
+\lambda_2(x-\mu)^2.
$$

Taking the exponential,

$$
p(x)
====

A\exp\left[
\lambda_1(x-\mu)
+\lambda_2(x-\mu)^2
\right],
$$

where

$$
A=e^{\lambda_0-1}.
$$

---

## Step 3: Apply the Mean Constraint

Because the distribution has mean $\mu$, it must be symmetric around $\mu$.

Therefore, the linear term must be zero:

$$
\lambda_1=0.
$$

For the density to converge as $x\rightarrow\pm\infty$, we must have

$$
\lambda_2<0.
$$

Write

$$
\lambda_2=-a,
\qquad a>0.
$$

Then

$$
p(x)=A\exp\left[-a(x-\mu)^2\right].
$$

---

## Step 4: Find the Normalization Constant

Using

$$
\int_{-\infty}^{\infty}p(x),dx=1,
$$

we get

$$
A\int_{-\infty}^{\infty}
e^{-a(x-\mu)^2},dx=1.
$$

Using the Gaussian integral

$$
\int_{-\infty}^{\infty}e^{-ay^2},dy
===================================

\sqrt{\frac{\pi}{a}},
$$

we obtain

$$
A\sqrt{\frac{\pi}{a}}=1.
$$

Therefore,

$$
A=\sqrt{\frac{a}{\pi}}.
$$

Thus,

$$
p(x)
====

\sqrt{\frac{a}{\pi}}
e^{-a(x-\mu)^2}.
$$

---

## Step 5: Apply the Variance Constraint

For the density

$$
p(x)
====

\sqrt{\frac{a}{\pi}}
e^{-a(x-\mu)^2},
$$

the variance is

$$
\operatorname{Var}(X)=\frac{1}{2a}.
$$

Since the given variance is $\sigma^2$,

$$
\frac{1}{2a}=\sigma^2.
$$

Therefore,

$$
a=\frac{1}{2\sigma^2}.
$$

The normalization constant becomes

$$
A
=

\sqrt{
\frac{1}{2\pi\sigma^2}
}
=

\frac{1}{\sqrt{2\pi\sigma^2}}.
$$

Hence,

$$
\boxed{
p(x)
====

\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left[
-\frac{(x-\mu)^2}{2\sigma^2}
\right]
}
$$

Therefore,

$$
\boxed{
X\sim\mathcal{N}(\mu,\sigma^2)
}
$$

---

## Maximum Entropy Value

The entropy of this normal distribution is

$$
\boxed{
H(X)
====

\frac{1}{2}
\ln\left(2\pi e\sigma^2\right)
}
$$

Therefore, among all continuous probability distributions on $\mathbb{R}$ having mean $\mu$ and variance $\sigma^2$, the Gaussian distribution has the maximum entropy.


# Q. Multiclass Logistic Regression: Automatic Classification of Emails

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

# Maximum-Entropy Email Classifier

Let

$$
q(\mathbf{x})=P(Y=1\mid \mathbf{x}).
$$

Therefore,

$$
P(Y=0\mid\mathbf{x})=1-q(\mathbf{x}).
$$

Since the marginal probabilities $P(\mathbf{x})$ are fixed, we maximize the conditional entropy

$$
H(Y\mid X)
==========

-\sum_{\mathbf{x}}P(\mathbf{x})
\left[
q(\mathbf{x})\ln q(\mathbf{x})
+
(1-q(\mathbf{x}))
\ln(1-q(\mathbf{x}))
\right].
$$

The constraints are

$$
\sum_{\mathbf{x}}
P(\mathbf{x})x_1q(\mathbf{x})=0.05,
$$

$$
\sum_{\mathbf{x}}
P(\mathbf{x})x_2q(\mathbf{x})=0.15,
$$

and

$$
\sum_{\mathbf{x}}
P(\mathbf{x})x_3q(\mathbf{x})=0.16.
$$

---

## Step 1: Construct the Lagrangian

Introduce Lagrange multipliers $\lambda_1,\lambda_2,\lambda_3$.

$$
\begin{aligned}
\mathcal{L}
=&H(Y\mid X)\
&+\lambda_1
\left[
\sum_{\mathbf{x}}P(\mathbf{x})x_1q(\mathbf{x})-0.05
\right]\
&+\lambda_2
\left[
\sum_{\mathbf{x}}P(\mathbf{x})x_2q(\mathbf{x})-0.15
\right]\
&+\lambda_3
\left[
\sum_{\mathbf{x}}P(\mathbf{x})x_3q(\mathbf{x})-0.16
\right].
\end{aligned}
$$

Differentiate with respect to $q(\mathbf{x})$:

$$
\frac{\partial\mathcal{L}}
{\partial q(\mathbf{x})}
========================

P(\mathbf{x})
\left[
\ln\left(
\frac{1-q(\mathbf{x})}{q(\mathbf{x})}
\right)
+
\lambda_1x_1+\lambda_2x_2+\lambda_3x_3
\right].
$$

At the maximum,

$$
\frac{\partial\mathcal{L}}
{\partial q(\mathbf{x})}=0.
$$

Therefore,

$$
\ln\left(
\frac{1-q(\mathbf{x})}{q(\mathbf{x})}
\right)
=======

-\left(
\lambda_1x_1+\lambda_2x_2+\lambda_3x_3
\right).
$$

Hence,

$$
\boxed{
P(Y=1\mid\mathbf{x})
====================

\frac{
\exp(\lambda_1x_1+\lambda_2x_2+\lambda_3x_3)
}{
1+\exp(\lambda_1x_1+\lambda_2x_2+\lambda_3x_3)
}
}
$$

or equivalently,

$$
\boxed{
P(Y=1\mid\mathbf{x})
====================

\frac{1}{
1+\exp[-(\lambda_1x_1+\lambda_2x_2+\lambda_3x_3)]
}
}
$$

This is the logistic maximum-entropy model.

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



# **Maximum Entropy (MaxEnt) Problem Formulation and Solution**

To find the least-biased probability density function $p(S_T)$ for the future asset price $S_T > 0$, we maximize the continuous Shannon entropy subject to our observed market constraints.

---

### **Step 1: Set Up the Optimization Objective and Constraints**

The objective is to maximize the Shannon entropy $H(p)$:

$$H(p) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T$$

Subject to the following four constraints:
1. **Total Probability (Normalization):**
$$\int_{0}^{\infty} p(S_T) \, dS_T = 1$$
2. **Risk-Free Growth (Forward Price):**
$$\int_{0}^{\infty} S_T \, p(S_T) \, dS_T = 102$$
3. **Market Volatility (Second Moment):**
$$\int_{0}^{\infty} S_T^2 \, p(S_T) \, dS_T = 10,600$$
4. **Asymmetric Market Sentiment (Skewness Penalty):**
$$\int_{0}^{\infty} \ln(S_T) \, p(S_T) \, dS_T = 4.60$$

---

### **Step 2: Construct the Lagrangian Function**

We introduce Lagrange multipliers $\lambda_0, \lambda_1, \lambda_2, \lambda_3$ for each respective constraint to form the Lagrangian functional $\mathcal{L}$:

$$\begin{aligned}\mathcal{L}(p, \lambda_0, \lambda_1, \lambda_2, \lambda_3) = & -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T \\ & - \lambda_0 \left( \int_{0}^{\infty} p(S_T) \, dS_T - 1 \right) \\ & - \lambda_1 \left( \int_{0}^{\infty} S_T \, p(S_T) \, dS_T - 102 \right) \\ & - \lambda_2 \left( \int_{0}^{\infty} S_T^2 \, p(S_T) \, dS_T - 10,600 \right) \\ & - \lambda_3 \left( \int_{0}^{\infty} \ln(S_T) \, p(S_T) \, dS_T - 4.60 \right)\end{aligned}$$

---

### **Step 3: Take the Functional Derivative**

To find the functional form maximizing $\mathcal{L}$, we take the directional derivative (Euler-Lagrange equation) with respect to $p(S_T)$ and set it to zero:

$$\frac{\delta \mathcal{L}}{\delta p(S_T)} = - \ln p(S_T) - 1 - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T) = 0$$

---

### **Step 4: Solve for the Analytical Distribution Form**

Isolating the log-density term:

$$\ln p(S_T) = -(1 + \lambda_0) - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T)$$

Taking the exponential of both sides yields the functional form of our distribution:

$$p(S_T) = \exp(-(1 + \lambda_0)) \cdot \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right) \cdot \exp\left(-\lambda_3 \ln(S_T)\right)$$

Using the algebraic property $\exp(-\lambda_3 \ln(S_T)) = S_T^{-\lambda_3}$, we simplify the expression into:

$$p(S_T) = \frac{1}{Z} S_T^{-\lambda_3} \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right)$$

Where $Z = \exp(1 + \lambda_0)$ serves as the **partition function (normalization constant)** defined by:

$$Z(\lambda_1, \lambda_2, \lambda_3) = \int_{0}^{\infty} S_T^{-\lambda_3} \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right) dS_T$$

---

### **Step 5: Mathematical Interpretation of the Result**

The derived distribution belongs to the **exponential family** and generalizes several standard distributions:
* If $\lambda_2 = 0$ and $\lambda_3 = 0$, it reduces to an **Exponential Distribution**.
* If $\lambda_3 = 0$ and $\lambda_1, \lambda_2 \neq 0$, it represents a **Truncated Normal Distribution** on $(0, \infty)$.
* With all parameters active ($\lambda_1, \lambda_2, \lambda_3 \neq 0$), it yields a highly flexible, skewed, fat-tailed distribution capable of perfectly accommodating the asymmetric risk-neutral market constraints without any pre-imposed Gaussian restrictions.

*(Note: In practice, finding the exact numerical values for $\lambda_1, \lambda_2, \lambda_3$ requires solving the dual problem numerically via non-linear optimization algorithms like L-BFGS or Newton-Raphson to match the target aggregates 102, 10,600, and 4.60).*
